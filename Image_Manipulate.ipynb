```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
im=cv2.imread('ttd.png')
cv2_imshow(im)
im.shape
```

```python
im2=cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
#im2=cv2.transpose(im2)
cv2_imshow(im2)
print(im2.shape)
```

**compress**

```python
newdim=(36,111)
im2=cv2.resize(im2, newdim, interpolation = cv2.INTER_AREA)
cv2_imshow(im2)
```

```python

dim1=im2.shape
size=dim1[0]*dim1[1]
dim2=(1,size)
data=cv2.resize(im2, dim2)
print(data.shape)
data
```

**Retrieve from database**

```python
dim3=(dim1[1],dim1[0])
data2=cv2.resize(im2, dim3)
#data2=cv2.transpose(data2)
cv2_imshow(data2)
print(data2.shape)
```
